```python
def trans(x):
    return np.exp(x)/(np.exp(x)+1)

gt = pd.read_csv('g.csv',header = None)
gt.index = trans(np.linspace(-18,18,1001))
gt.columns = np.linspace(-1,1,201)
g_R = gt.sum(axis=1)*0.01
g_Z = gt.sum(axis=0)*0.036
plt.plot(g_R)

gt = pd.read_csv('g_1.csv',header = None)
gt.index = trans(np.linspace(-18,18,1001))
gt.columns = np.linspace(-1,1,201)
g_R2 = gt.sum(axis=1)*0.01
g_Z2 = gt.sum(axis=0)*0.036
plt.plot(g_R2)
```

```python
def read_csv(name):
    h1 = pd.read_csv(name+'.csv',header = None)
    h1.index = trans(np.linspace(-18,18,1001))
    h1.columns = np.linspace(-1,1,201)
    return h1
```

```python
h1 = read_csv('h1')
h2 = read_csv('h2')
hz = read_csv('hz')
d1 = read_csv('d1')
d2 = read_csv('d2')
h12 = read_csv('h12')
h22 = read_csv('h22')
hz2 = read_csv('hz2')
d12 = read_csv('d12')
d22 = read_csv('d22')
```

```python
fig, axes = plt.subplots(2,2,figsize = (12,8))
axes = axes.flatten()
sns.lineplot(d1.iloc[:-10,100], ax = axes[0], label = 'Z = 0', color = 'red')
sns.lineplot(d2.iloc[10:,100], ax = axes[1], label = 'Z = 0', color = 'red')
sns.lineplot(d1.iloc[:-10,127], ax = axes[0], label = r'$Z =  +1.64\sigma$', color = 'blue')
sns.lineplot(d2.iloc[10:,127], ax = axes[1], label = r'$Z =  +1.64\sigma$', color = 'blue')
sns.lineplot(d1.iloc[:-10,73], ax = axes[0], label = r'$Z =  -1.64\sigma$', color = 'green')
sns.lineplot(d2.iloc[10:,73], ax = axes[1], label = r'$Z =  -1.64\sigma$', color = 'green')

sns.lineplot(h1[0], ax = axes[2], label = '$-H_1$', color = 'red')
sns.lineplot(h2[0], ax = axes[2], label = '$-H_2$', color = 'blue')
sns.lineplot(hz[0], ax = axes[2], label = '$-H_z$', color = 'green')

sns.lineplot(g_R, ax = axes[3], color = 'red')


axes[0].set_title(r'${I^1}/{K^1}$')
axes[1].set_title(r'${I^2}/{K^2}$')
axes[2].set_title(r'$-H$')
axes[3].set_title('R stationary density')

# axes[0,1].set_title('I_1/K_1')
# axes[1,0].set_title('I_2/K_2')
# axes[1,1].set_title(r'h')
axes[0].set_ylim([0.03,0.036])
axes[1].set_ylim([0.03,0.036])
axes[2].set_ylim([-0.01,0.18])
# axes[1,1].set_ylim([-0.01,0.09])

for ax in axes:
    ax.set_ylabel('')
# fig.suptitle('Symmetric Case')
fig.tight_layout()
fig.savefig('No titile Symmtric Case.png', dpi = 1200)
plt.show()
```
